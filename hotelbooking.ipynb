# Data Analysis and Visualizations

```python
cancel_per = df['is_canceled'].value_counts(normalize=True)
```

```python
plt.bar(['Not_canceled','Canceled'],df['is_canceled'].value_counts(),width=0.5)
```

```python
# Convert 'is_canceled' to string type
df['is_canceled'] = df['is_canceled'].astype(str)
```

```python
plt.figure(figsize=(8,4))
ax1 = sns.countplot(x='hotel',hue='is_canceled',data=df,palette='Blues')
plt.title('Reservation status in different hotels')
plt.xlabel('hotel')
plt.ylabel('number of reservation')
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (20,8))
plt.title('Average daily rate of city and resort hotel',fontsize = 30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize=(20,10))
sns.countplot(x='month',hue='is_canceled',data=df,palette='bright')
plt.legend(['not canceled','canceled'])
plt.title('Reservation status per month',fontsize=30)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Check if there is any data in the filtered DataFrame
filtered_df = df[df['is_canceled'] == 1]
if filtered_df.empty:
    print("No data found for canceled bookings.")
else:
    # Group by month and sum the adr values
    grouped_data = filtered_df.groupby('month')[['adr']].sum().reset_index()

    # Check if grouped data is empty
    if grouped_data.empty:
        print("Grouped data is empty.")
    else:
        # Plot the barplot
        sns.barplot(x='month', y='adr', data=grouped_data)
        plt.title('ADR per month', fontsize=20)
        plt.show()
```
